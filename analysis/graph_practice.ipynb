```python
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
iris=px.data.iris()# iris is a database and after dot it is a pandas liberary
iris.columns.tolist()
```

```python
iris
```

```python
iris.sepal_length
```

```python
iris.sepal_length.plot()
```

```python
#theme change in matplotlib 
themes=plt.style.available
print(themes)
print(len(themes))
```

```python
for theme in themes:
    plt.style.use(theme)
    iris.sepal_length.plot(title=f'{theme}theme')
    plt.show()
```

```python
plt.style.use('seaborn-v0_8-bright')
```

```python
iris.sepal_length.plot(figsize=(15,5))#figsize is used to define the size of graph
```

```python
iris.sepal_length.plot(figsize=(15,5),color='red',marker='o',markersize='13',alpha=.5)#here alpha is translucent
```

```python
iris[['sepal_length','sepal_width']].plot(figsize=(15,5))
```

```python
iris[['sepal_length','sepal_width']].plot(figsize=(10,5),subplots=True)
```

```python
iris[['sepal_length','sepal_width']].plot(figsize=(15,5),subplots=True,layout=(1,2))
```

```python
cols=['sepal_length','sepal_width','petal_length','petal_width']
iris[cols].plot(figsize=(15,5),subplots=True,layout=(2,2))
```

```python
iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
```

```python
iris.plot(kind='scatter',
x='sepal_length',
y='sepal_width',
c='species_id',
colormap='rainbow',
s=100)
```

```python
iris.plot(kind='hexbin',x='sepal_length',y='sepal_width',gridsize=10)
```
