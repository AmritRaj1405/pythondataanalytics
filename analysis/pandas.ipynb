```python
#for merging data
a=pd.DataFrame({
    'name':['Ajay','Arvind','Bhavesh','Binay'],
    'class':[10,9,10,11],
})
b=pd.DataFrame({
    'science':[90,78,89,67],
    'maths':[78,87,65,76],
    'english':[90,56,78,88],
})
```

```python

a=pd.DataFrame({
    'name':['Ajay','Arvind','Bhavesh','Binay'],
    'class':[10,9,10,11],
    'id':[1,2,3,4]
})
b=pd.DataFrame({
    'science':[90,78,89,67,90],
    'maths':[78,87,65,76,77],
    'english':[90,56,78,88,66],
    'id':[1,2,3,5,4]
})
```

```python
out=a.merge(b,how='inner',on='id')
out
```

```python
out=a.merge(b,how='outer',on='id')
out
```

```python
out=a.merge(b,how='left',on='id')
out
```

```python
out=a.merge(b,how='right',on='id')
out
```

```python
canada=pd.read_excel('canada.xlsx',sheet_name='Canada by Citizenship',skiprows=20,skipfooter=2)
```

```python
canada.set_index('OdName',inplace=True)
```

```python
canada
```

```python
year=list(range(1980,2014))
print(year)
```

```python
canada[year]
```

```python
canada.loc[['India','China'],year].T.plot(kind='line',figsize=(15,5))
```

```python
pd.set_option('display.max_column',None)
```

```python
canada['Total']=canada[year].sum(axis=1)
```

```python
canada.sort_values(by='Total',ascending=False,inplace=True)
canada.head()
```

```python
canada.groupby('AreaName').groups 
```

```python
canada.groupby('AreaName').get_group('Asia')
```

```python
canada.groupby('AreaName')['Total'].sum()
```

```python
canada.groupby('AreaName')[year].sum().T.plot(figsize=(15,5),kind='area',alpha=0.5)
```

```python
canada.groupby('DevName').groups
```

```python
canada.groupby('DevName')[year].sum().T.plot(figsize=(15,5),kind='area',alpha=0.5)
```

```python
canada.groupby('DevName')['Total'].sum().plot(kind='pie',figsize=(5,5),autopct='%1.1f%%')
```

```python
canada.groupby('DevName')['Total'].agg(['mean','sum','count','min','max','std'])
```

```python
canada.groupby('AreaName')['Total'].agg(['mean','max','min','count','std'])
```

```python
_=canada.groupby(['AreaName','RegName'])[year].sum().T.plot(subplots=True,layout=(12,3),figsize=(15,50),kind='area',grid=True)
```

```python
cols_to_drops=['Type','Coverage','AREA','REG','DEV']
canada.drop(columns=cols_to_drops, inplace=True)
```

```python
canada.head()
```

```python
canada.rename(
    {'AreaName':'Continent','RegName':'Region','DevName':'Status'},
    axis=1,
    inplace=True
)
```

```python
canada.rename(
    {'United Kingdom of Great Britain and Northern Ireland':'UK',},
    inplace=True
)
```

```python
canada.head()
```

```python
canada.to_csv('canada.csv')
```
